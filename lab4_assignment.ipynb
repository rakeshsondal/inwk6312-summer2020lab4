```python
import string
import math
document=open("wordfile.txt",encoding="utf-8")
new_doc=[]
L={}
count=0
for i in document:
    for j in i.split():
        n=j.strip(string.punctuation).lower()
        new_doc.append(n)
        if n not in new_doc:
            count=1
            L.update
        else:
            count=count+1
            L.update

print(new_doc)
print(len(new_doc))
print("Number of times each word is used: ",L)
```
